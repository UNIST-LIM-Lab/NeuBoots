```python
import numpy as np
from utils.jupyter import plot_multiclass_calibration_curve
from scipy.special import softmax
from sklearn.calibration import calibration_curve
```

```python
cifar100_base = np.load('outs/cutout/cifar100/output.npy')[0]
mp0, acc0, mc0, cnt0 = plot_multiclass_calibration_curve(cifar100_base[:, :-1], cifar100_base[:, -1], title='CIFAR-100 Base')
```

```python
cifar100 = np.load('outs/cutout/cifar100/output.npy')
mp0, acc0, mc0, cnt0 = plot_multiclass_calibration_curve(softmax(cifar100[:, :, :-1], -1).mean(0), cifar100[0, :, -1], title='CIFAR-100', is_softmax=True)
```

```python
conf = softmax(cifar100[..., :-1], -1)
label = cifar100[0, :, -1]

mean = conf.mean(0)
# std = np.repeat(conf.std(0).mean(1)[:, None], 100, 1)
std = conf.std(0)
mp0, acc0, mc0, cnt0 = plot_multiclass_calibration_curve((mean-std*2).clip(0, 1), label, is_softmax=True, title='CIFAR-100 Lower Bound')
```

```python
conf = softmax(cifar100[..., :-1], -1)
label = cifar100[0, :, -1]

mean = conf.mean(0)
# std = np.repeat(conf.std(0).mean(1)[:, None], 100, 1)
std = conf.std(0)
mp0, acc0, mc0, cnt0 = plot_multiclass_calibration_curve((mean+std*2).clip(0, 1), label, is_softmax=True, title='CIFAR-100 Upper Bound')
```
